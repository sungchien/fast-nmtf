```python
#!/usr/bin/env python

import numpy as np
import pandas as pd
import csv
from sklearn.metrics import pairwise_distances
from sklearn.cluster import AgglomerativeClustering
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
card_data = np.load('card_data.npz')
```

```python
fp = open("U.csv")
reader = csv.reader(fp)
card_vector = []
for row in reader:
    if len(row) > 0:
        card_vector.append(row)
fp.close()
```

```python
pairwise_distances(np.array(card_vector), metric='cosine')
```

```python
X_embedded = TSNE(n_components=2, metric='cosine').fit_transform(np.array(card_vector))
```

```python
X_embedded
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(X_embedded[:,0], X_embedded[:,1])

for i, txt in enumerate(card_data['cards']):
    ax.annotate(i, (X_embedded[i, 0], X_embedded[i, 1]))
```
