```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
err_data = pd.read_csv('card_sorting_err.csv')
err_data.shape
```

```python
err_data.head()
```

```python
f, ax = plt.subplots(nrows= 10, ncols=2, figsize=(14, 20))

for i in range(1, 21, 1):
    rix = int(np.floor((i-1)/2))
    cix = (i-1)%2
    sns.lineplot(x="k", y="err", data=err_data[err_data.seed==i], ax=ax[rix][cix])
    ax[rix][cix].set_xticks(range(1, 21, 1))
```
